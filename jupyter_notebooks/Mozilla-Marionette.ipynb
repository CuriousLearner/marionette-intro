```python
from marionette import Marionette
```

```python
client = Marionette('localhost', port=2828)
```

```python
client.start_session()
```

```python
client.session
```

```python
client.set_context(client.CONTEXT_CHROME)
```

```python
client.page_source
```

```python
client.set_context(client.CONTEXT_CONTENT)
```

```python
client.page_source
```

```python
client.navigate('http://www.mozilla.org')
```

```python
client.navigate('http://www.SanyamKhurana.com/#/contact')
```

```python
client.go_back()
```

```python
client.go_forward()
```

```python
from marionette_driver import By
from marionette_driver.marionette import HTMLElement
```

```python
first_name = client.find_element(By.ID, 'first_name')
```

```python
last_name = client.find_element(By.ID, 'last_name')
```

```python
email = client.find_element(By.ID, 'email')
```

```python
message = client.find_element(By.ID, 'message')
```

```python
first_name.send_keys("Sanyam")
last_name.send_keys("Khurana")
email.send_keys("Sanyam@SanyamKhurana.com")
message.send_keys("Hey there! I love Python and Mozilla!")
```

```python
button = client.find_element(By.NAME, 'action')
```

```python
button.click()
```

```python
button.get_attribute('class')
```

```python
# Executing JS :)

client.execute_script("""return arguments[0] * arguments[1]; """, script_args=[5, 7])
```

```python
client.execute_script("alert('I love Mozilla!')")
```
